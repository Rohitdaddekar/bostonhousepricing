```python
# Dependent and independent features
X = dataset.iloc[:,:-1]
X
```

```python
y = dataset.iloc[:,-1]
y
```

```python
#train-test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
X_train= sc.fit_transform(X_train)
X_train
```

```python
X_test= sc.transform(X_test)
X_test
```

# Model training

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
```

```python
regression.fit(X_train,y_train)
```

```python
regression.coef_
```

```python
regression.intercept_
```

```python
# check on which parameter our model is trained
regression.get_params()
```

```python
reg_pred = regression.predict(X_test)
reg_pred
```

```python
# plot a scatter plot for predication
plt.scatter(y_test,reg_pred)
```

```python
#residual
residual= y_test-reg_pred
residual
```

```python
#plot this residuals
sns.displot(residual,kind="kde")
```

```python
#scatter plot wrt predication and residuals
#its uniformly distributated
plt.scatter(reg_pred,residual)
```

```python
#checking the performance metrix of model
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(mean_squared_error(y_test,reg_pred))
print(mean_absolute_error(y_test,reg_pred))
print(np.sqrt(mean_squared_error(y_test,reg_pred)))
```

# R2 and adjusted R2

```python
from sklearn.metrics import r2_score
score= r2_score(y_test,reg_pred)
score
```

```python
#adjusted R2 value:
adj_R2 = 1-(1-score)*(len(y_test)-1)/(len(y_test)-(X_test.shape[1]-1))
adj_R2
```
